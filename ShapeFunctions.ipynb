```python
import numpy as np
import matplotlib.pyplot as pl
from mpl_toolkits.mplot3d import Axes3D

# you can replace inline with qt to get plots in a separate window
%matplotlib inline
```

```python
#Quadrilateral element:
x=np.linspace(0,1,21)
y=np.linspace(0,1,21)
z=np.zeros((21, 21))
for i in range(len(x)):
    for j in range(len(y)):
        z[i,j]=(1/4)*(x[i]-1)*(y[j]-1)

# Create figure and increase spacing between subplots
pl.figure(1, figsize=(9.75, 8))
pl.subplots_adjust(wspace=0.25)

# plot contours of quadrilateral element shape function
#pl.subplot(121)
pl.contourf(x, y, z, 24, cmap="Reds")
pl.colorbar(label="$N_1$")

#pl.grid()
pl.axis('equal')
pl.xlabel("x")
pl.ylabel("y")
pl.title("Quadrilateral Element Shape Function")
```

```python
#Beam element:
xB=np.linspace(0,1,21)
HC1=np.zeros((21))
HC2=np.zeros((21))
HC3=np.zeros((21))
HC4=np.zeros((21))

HC1[:]=1-3*xB[:]**2+2*xB[:]**3
HC2[:]=xB[:]-2*xB[:]**2+xB[:]**3
HC3[:]=3*xB[:]**2-2*xB[:]**3
HC4[:]=-xB[:]**2+xB[:]**3


# Create figure and increase spacing between subplots
pl.figure(1, figsize=(20, 5))
pl.subplots_adjust(wspace=0.25)

pl.subplot(141)
pl.plot(xB, HC1, color="r")
pl.grid()
pl.axis('equal')
pl.xlabel("x")
pl.ylabel("y")
pl.title("Hermite Cubic Function 1")

pl.subplot(142)
pl.plot(xB, HC2, color="r")
pl.grid()
pl.axis('equal')
pl.xlabel("x")
pl.ylabel("y")
pl.title("Hermite Cubic Function 2")

pl.subplot(143)
pl.plot(xB, HC3, color="r")
pl.grid()
pl.axis('equal')
pl.xlabel("x")
pl.ylabel("y")
pl.title("Hermite Cubic Function 3")

pl.subplot(144)
pl.plot(xB, HC4, color="r")
pl.grid()
pl.axis('equal')
pl.xlabel("x")
pl.ylabel("y")
pl.title("Hermite Cubic Function 4")

pl.suptitle("Beam Element Shape Functions")
```

```python
# Create figure and increase spacing between subplots
pl.figure(1, figsize=(8, 8))
pl.plot(xB, HC1, label='HC1')#, color="r")
pl.plot(xB, HC2, label='HC2')#, color="r")
pl.plot(xB, HC3, label='HC3')#, color="r")
pl.plot(xB, HC4, label='HC4')#, color="r")
pl.grid()
pl.xlim([0,1])
pl.ylim([-0.2,1])
pl.axis('equal')
pl.xlabel("x")
pl.ylabel("y")

pl.legend(loc=7)

pl.show()

#100 143 255
#120 94 240
#220 38 127
#254 97 0
#255 176 0
```

```python
#Truss element:
xB=np.linspace(0,1,21)
SF1=np.zeros((21))
SF2=np.zeros((21))

SF1[:]=1-xB[:]
SF2[:]=xB[:]


# Create figure and increase spacing between subplots
pl.figure(1, figsize=(10, 5))
pl.subplots_adjust(wspace=0.25)

pl.subplot(121)
pl.plot(xB, SF1, color="r")
pl.grid()
pl.axis('equal')
pl.xlabel("x")
pl.ylabel("y")
pl.title("Shape Function 1")

pl.subplot(122)
pl.plot(xB, SF2, color="r")
pl.grid()
pl.axis('equal')
pl.xlabel("x")
pl.ylabel("y")
pl.title("Shape Function 2")

pl.suptitle("Truss Element Shape Functions")
```

```python
# Create figure and increase spacing between subplots
pl.figure(1, figsize=(8, 8))
pl.plot(xB, SF1, label='SF1')#, color="r")
pl.plot(xB, SF2, label='SF2')#, color="r")
pl.grid()
pl.xlim([0,1])
pl.ylim([-0.2,1])
pl.axis('equal')
pl.xlabel("x")
pl.ylabel("y")

pl.legend(loc=7)

pl.show()

#100 143 255
#120 94 240
#220 38 127
#254 97 0
#255 176 0
```
